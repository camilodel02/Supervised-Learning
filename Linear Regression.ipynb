```python
x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))  # Normalización de x entre 0 y 1
y_norm = (y - np.min(y)) / (np.max(y) - np.min(y))
display(x_norm,y_norm)
```

## Compute cost

```python
def compute_cost(x,y,w,b):
    m=x.shape[0] #Toma la cantidad de filas del vector de x (cantidad de features)
    cost =0

    for i in range(m):
        f_wb=w*x[i]+b 
        cost=cost+(f_wb-y[i])**2 #total error
        total_cost=1/(2*m)*cost #Average error
    return total_cost
```

## Compute gradient

```python
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0

    for i in range(m):
        f_wb=w*x[i]+b #model 
        dj_dw_i=(f_wb-y[i])*x[i] #partial derivative of J(w,b) respect to w
        dj_db_i=f_wb-y[i] #partial derivative of J(w,b) respect to b
          #sum of dj_w for each feature
        dj_db+=dj_db_i
        dj_dw+=dj_dw_i  #sum of dj_b for each feature
    dj_dw=dj_dw/m  #1/m term multiplpy by dj_dw
    dj_db=dj_db/m   #1/m term multiplpy by dj_db

    return dj_dw,dj_db
```

## Gradient descent algorithm

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function):

    J_history=[]
    p_history=[]
    b=b_in
    w=w_in

    for i in range(num_iters):
        dj_dw, dj_db =gradient_function(x, y, w, b)

        b=b-alpha *dj_db
        w=w-alpha*dj_dw

        if i<100000:
            J_history.append(cost_function(x, y, w, b))
            p_history.append([w, b])
        if i% math.ceil(num_iters/10)==0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e}",
                  f"dj_dw: {dj_dw:0.3e}, dj_db: {dj_db:0.3e}",
                  f"w: {w:0.3e}, b:{b:0.5e}")
    return w, b, J_history, p_history 
```

```python
w_init=0
b_init=0

iterations = 10000
tmp_alpha=0.001

w_final, b_final, J_hist, p_hist =gradient_descent(x_norm, y_norm, w_init, b_init, tmp_alpha,
                                                   iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```
